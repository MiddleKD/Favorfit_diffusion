```python
from PIL import Image
img = Image.open("./images/demo_image/cocktail.jpg").convert("RGB")
```

```python
img_1 = img.resize([512,512])
img_2 = img.resize([512,700])
img_3 = img.resize([700,512])
```

```python
import torch
from diffusers import StableDiffusionImg2ImgPipeline

device = "cuda"
model_id_or_path = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionImg2ImgPipeline.from_pretrained(model_id_or_path, torch_dtype=torch.float16)
pipe = pipe.to(device)
```

```python
prompt = "A blue cocktailin the cave"

images = pipe(prompt=prompt, 
              image=img_3, 
              strength=1.0, 
              guidance_scale=7.5, 
              num_inference_steps=20).images

print(images[0].size)
images[0]
```
